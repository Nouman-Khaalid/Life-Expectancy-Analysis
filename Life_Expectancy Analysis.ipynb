```python
nomi_df=pd.read_csv('E:\Jupyter Notebook\\Life-Expectancy-Data-Updated.csv')
nomi_df
```

Question 1:Average life expectancy across all countries:

```python
average_life_expectancy = nomi_df['Life_expectancy'].mean()
average_life_expectancy
```

```python



#  'Life_expectancy' for one country (e.g., Turkey) yearly
turkey_data = nomi_df[nomi_df['Country'] == 'Turkiye']

plt.figure(figsize=(8, 5))
plt.plot(turkey_data['Year'], turkey_data['Life_expectancy'], marker='o', color='b', linestyle='-')
plt.title('Life Expectancy Over Years in Turkey')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.grid(True)
plt.show()
```

```python


plt.figure(figsize=(8, 5))
sns.histplot(nomi_df['Life_expectancy'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Life Expectancy')
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```
